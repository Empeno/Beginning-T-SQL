# Arbejde med NULL

  

* Arbejdet med NULL er måske største udfordring
* NULL betyder at der ikke er angivet en værdi for en bestemt kolonne i en række
* Valgfrit telefonnummer i købsflow
* Det kan betyde ingen telefon eller privacy
* Resultat: Du kender ikke telefonnummer
* Ofte laver vi filtrering baseret på kendskab til værdi
* NULL udfordrer denne form for filtrering
* Overvej: tabel med 1.000.000 telefonnumre -> find kunde

```sql
USE AdventureWorks2019
SELECT MiddleName
FROM Person.Person;
```

```sql
SELECT MiddleName
FROM Person.Person
WHERE MiddleName = 'B';
```

```sql
SELECT MiddleName
FROM Person.Person
WHERE MiddleName != 'B';
```

- Forespørgsel 1 uden WHERE-sætning returnerer 19.972 rækker
    
- Forespørgsel 2 returnerer 291 rækker med mellemnavnet B
    
- Logisk følger, at forespørgsel 3 vil returner forskellen på de to tal: 19.681 rækker.
    
- Når du kontrollerer resultaterne af forespørgsel 3: der ikke tages højde for mere end 8000 rækker. 
    
- **Det er fordi rækkerne med manglende mellemnavne ikke kan findes af udtrykket != (NOT EQUAL)**
    
- Sammenligning af NULL med B returnerer UNKNOWN, så rækkerne returneres ikke.

* Normalt sammenligner du dataene i en kolonne med en værdi - TRUE eller FALSE
* eller sammenligner om værdierne fra to kolonner returnerer enten TRUE eller FALSE.
* Hvis udtrykket i WHERE-klausulen evalueres til TRUE, så returneres rækken.
* Hvis udtrykket evalueres til FALSE, returneres rækken ikke
* Hvis en værdi i udtrykket indeholder NULL resulterer udtrykket i UNKNOWN.
* På nogle måder er adfærden som FALSE: Når et resulterer UNKNOWN returneres rækken ikke. 
* Problemerne begynder, når du bruger enhver anden operator undtagen end lig med (=).
* Det modsatte af FALSE er TRUE, men det modsatte af UNKNOWN er stadig UNKNOWN.

## Bemærk

- NULL ikke er en værdi, så den ikke kan være lig med UNKNOWN eller anden værdi.
- T-SQL giver i stedet specifikke udtryk og funktioner til test for NULL værdier.
- For at teste for en NULL skal du bruge IS NULL eller IS NOT NULL.
- Derudover kan du ændre en NULL til en værdi som f.eks. en tom streng ('').
- Glemmer du at tage mulige NULL'er i betragtning -> ufuldstændige resultater.
- **Husk altid at tænke på NULL**
- **især når du skriver et hvilket som helst udtryk indeholdende NOT**
- Tilhører NULL'erne resultaterne? I så fald bliver du nødt til at kontrollere for NULL.
- Hold også NULL i tankerne, ved brug af mindre end (<). 
- NULL vil også blive udeladt af disse resultater. 
- En forespørgsel, der kontrollerer NULL, ser sådan ud

```sql
SELECT MiddleName
FROM Person.Person
WHERE MiddleName IS NULL;
```

- Forstå NULL påvirker resultaterne af queries: en af vigtigste færdigheder
- Selv erfarne T-SQL-udviklere kæmper med NULL

```sql
SELECT FirstName, MiddleName
FROM Person.Person
WHERE MiddleName IS NULL
 OR MiddleName !='B';
```
